# 2. Bernoulli distribution

### Ice-cream example part 2

The toddler has an 8 year old sibling who has been promised that they can order sprinkles if they have been well-behaved. Unfortunately, they do not know if their parents are aware that they tidied up their room by stuffing everything under the bed. They estimate that they were very sneaky in doing so, and that the probability that their parents believe they have been good as 0.8.

Write down the probability distribution for getting sprinkles. What is the variance of this distribution?

<details> <summary>▶️ Click to show the solution</summary>

Let $S=1$ represent the child getting sprinkles, and $S=0$ represent them missing out. Since there are only two outcomes, $S$ can be modelled as Bernoulli random variable with parameter $p=0.8$.

For the Bernoulli distribution, we know $\text{Var}(S) = p(1-p) = 0.8 * 0.2 = 0.16$.

</details>

# 3. Binomial distribution

### Icecream example part 3
While you are eating your icecream (coffee with chocolate sauce on top), you are suprised to notice multiple people ordering vanilla. You estimate that the probability of ordering vanilla is 0.36.

As you continue to eat your icecream, you observe 20 more customers ordering. Let $X$ be the random variable representing whether they order vanilla.

a) What is the distribution of $X$? Write down the pmf of $X$.

b) What is the probability that exactly 4 people order vanilla? What about more than 17 people?

c) What is the expected number of vanilla orders?

````r
choose(20,18)*0.36^18*0.64^2 + choose(20,19)*0.36^19*0.64^1 + choose(20,20)*0.36^20*0.64^0
````

<details> <summary>▶️ Click to show the solution</summary>

a) Let $X$ be the number of customers who order vanilla. Then $X \sim \text{Bin}(20, 0.36)$. It has pmf:
$$Pr(X = k) = \binom{20}k (0.36)^k (0.64)^{20-k} \quad \text{for } k=0,\ldots,20$$

b) We need to compute $Pr(X=4)$.

$$P(X=4) = \binom{20}4 (0.36)^4 (0.64)^{16} \approx 0.0645$$

Now for the probability that more than 17 people order vanilla:
\begin{align*} P(X > 17) & = P(X = 18) + P(X=19) + P(X=20) \\
& = \binom{20}{18} 0.36^{18} 0.64^2 + \binom{20}{19} 0.36^{19} 0.64 + \binom{20}{20} 0.36^{20} \\
& \approx 8.52 \times 10^{-7}
\end{align*}

c) For the binomial distribution, $\mathbb{E}(X) = np = 20 \cdot 0.36 = 7.2$, so the expected number of vanilla orders is 7.2.
</details>

## Binomial random variables in R


````r
dbinom(4, size=20, prob=0.36)
# P(X > 17) = 1 - P(X <= 17)
1 - pbinom(10, size=20, prob=0.36)
qbinom(0.75, size=20, prob=0.36)
````


#### **R Quick Reference**

`R` supports the following functions for computing distributional quantities and simulating from binomial distributions:

- `dbinom(x, size, prob, log = FALSE)` computes PMF of `Binomial(size,p)`;
- `pbinom(q, size, prob, lower.tail = TRUE, log.p = FALSE)` computes CDF of `Binomial(size,p)`;
- `qbinom(p, size, prob, lower.tail = TRUE, log.p = FALSE)` computes `p`-quantile of `Binomial(size,p)`;
- `rbinom(n, size, prob)` simulates `n` random numbers from `n` independent r.v. following `Binomial(size,p)`.

Here,

- `x`, `q`: The number of successes.
- `size`: The number of Bernoulli trials. If `size = 1`, we obtain the Bernoulli distribution.
- `prob`: The probability of success
- `log`: Whether or not we want to return log PMF.
- `lower.tail`: If `TRUE`, return $\Pr(X \leq x)$ where $X \sim \text{Binomial(size,p)}$. If `FALSE`, return $\Pr(X > x)$.

# 4. Geometric distribution

### Ice-cream example part 4
Customers order toppings independently with probability 0.4. Let $Y$ be the number of customers up till and including the first customer of the day who orders a topping.

a) Compute the probability that $Y=4$ and $Y \geq 3$.

b) Calculate $\mathbb{E}(Y)$ and interpret this value.

<details> <summary>▶️ Click to show the solution</summary>

We can model $Y$ as a geometric random variable with parameter $p = 0.4$, giving us the pmf $P(Y = k) = 0.6^{k-1} 0.4$ for $k \geq 1$.

a) $$Pr(Y = 4)  = 0.6^3 \cdot 0.4 = 0.0864$$

\begin{align*} Pr(Y \geq 3) & = 1 - Pr(Y \leq 2) = 1 - (Pr(Y=2) + Pr(Y=1)) \\
& = 1 - (0.6^1 \cdot 0.4 + 0.6^0 \cdot 0.4) \\
& = 1 - 0.64 = 0.36 \end{align*}

b) $$\mathbb{E}(Y) = \frac1p = \frac1{0.4} = 2.5$$
On average, we expect the first topping order to be from the second or third customer of each day.


</details>

## Geometric distribution in R

#### **R Quick Reference**

`R` supports the following functions for computing distributional quantities and simulating from geometric distributions:

- `dgeom(x, prob, log)` computes PMF of `Geometric(prob)`;  
- `pgeom(q, prob, lower.tail = TRUE)` computes CDF of `Geometric(prob)`;  
- `qgeom(p, prob)` computes `p`-quantile of `Geometric(prob)`;  
- `rgeom(n, prob)` simulates `n` random numbers from `Geometric(prob)

Importantly, the built-in geometric distribution in R counts the *number of failures* before success. If you need to calculate the total number of trials, you'll need to add 1 to `x`.

# 5. Negative binomial distribution

### Icecream example part 5
You are working at the icecream shop and notice that you are running low on chocolate flavour. You only have enough to make 2 more orders. If you run out, you will close the store to avoid customers becoming angry over the lack of chocolate. You estimate that the probability of a customer ordering chocolate is 0.27.

a) How many more customers do you expect to serve before you run out and close the store?

b) The store is only open for another 30 minutes anyway, during which period you estimate you can serve one customer per minute. What is the probability that you close the store at closing time without running out of chocolate first?

<details> <summary>▶️ Click to show the solution</summary>

Let $X$ be the number of customers until the second chocolate order. $X$ follows a negative binomial distribution with $k=2$ and $p=0.27$.

a) $$\mathbb{E}(X) = \frac{k}{p} = \frac{2}{0.27} \approx 7.41$$
You expect to serve 7 or 8 customers before running out of chocolate.

b) This time we can model the number of customers ordering chocolate as $Y$, a regular binomial distribution with 30 total customers and probability 0.27 of ordering chocolate. Then we want to find the probability that less than 2 out of 30 customers order chocolate.
\begin{align*}
Pr(Y < 2) & = P(Y=0) + P(Y=1) \\
& = (1-0.27)^30 + 30 \cdot 0.27 \cdot (1-0.27)^{29} \\
& \approx 0.00096
\end{align*}
So there is only a 0.096% chance that you make it to closing time without needing to close early because you ran out of chocolate.

</details>

## Question 5.1
The probability that a randomly selected box of a certain type of cereal has a particular prize is 0.2. Suppose you purchase box after box until you have obtained two of these prizes.


a) What is the probability that you purchase $x$ boxes until you obtain the two prizes?

b) What is the probability that you only have to purchase four boxes?

c) What is the probability that you have to purchase more than four boxes?

d) How many boxes without the desired prize do you expect to purchase?

<details> <summary>▶️ Click to show the solution</summary>

Let $X$ be the number of boxes we must purchase until we obtain two prizes. Then $X \sim \mathrm{NegBin}(2,0.2)$.


a) $$Pr(X = x) = \binom{x-1}{1} 0.2^2 0.8^{x - 2},\quad x \ge 2$$


b) \begin{align*}
 				Pr(X = 4) = \binom{4-1}{1} 0.2^2 0.8^{4-2} = 0.0768.
 			\end{align*}

c) \begin{align*}
				Pr(X > 4) &= 1 - Pr(X \le 4).\\
 				Pr(X \le 4) &= Pr(X = 2) + Pr(X = 3) + Pr(X = 4),\\
 					&= \binom{2-1}{1} 0.2^{2} 0.8^{2-2} + \binom{3-1}{1} 0.2^{2} 0.8^{3-2} + \binom{4-1}{1} 0.2^{2} 0.8^{4-2},\\
 					&= \binom{1}{1} 0.2^2 0.8^0 + \binom{2}{1} 0.2^2 0.8^1 + \binom{3}{1} 0.2^2 0.8^2,\\
 					&= 0.1808.\\
 				\therefore Pr(X > 4) &= 1 - 0.1808,\\
 					&= 0.8192
 			\end{align*}

d) We are interested in $\mathbb{E}(X - 2)$:
			\begin{align*}
 				\mathbb{E}(X-2) &= \mathbb{E}(X) - 2,\\
 					&= \dfrac{k}{p}  - 2 \quad \text{with } k=2,\:p=0.2,\\
 					&= \dfrac{2 }{0.2} - 2,\\
 					&= 8.
		 	\end{align*}
Alternatively, we could have parametrised differently, and let $Y$ be the *number of failures* until 2 successes. In this case, we would have used the other form of the negative binomial distribution for $Pr(Y = y)$, and in (d) we would look for $\mathbb{E}(Y)$.

</details>

# 6. Poisson distribution

### Ice-cream example part 6
Let $N$ be the number of toppings ordered in an hour.

a) Given an average rate of one customer per minute, and probability of ordering toppings of 0.4, calculate the average rate of toppings per hour. Compute $P(N = 15)$.  

b) We now consider a 15-minute interval. What is the corresponding rate parameter for toppings? What is the probability of no topping orders in this interval?



<details> <summary>▶️ Click to show the solution</summary>

We can model $N$ using a Poisson distribution.

a) Over the hour we expect $\lambda = 60 \times 0.4 = 24$ toppings/hour, and this becomes the parameter of our Poisson distribution, i.e. $N \sim \text{Poisson}(24)$.

$$Pr(N=15) = \frac{24^{15} e^{-24}}{15!} \approx 0.015$$

b) In a 15-minute interval, we expect $24 \times 0.25 = 6$ toppings/15 minutes (as 15 minutes is a quarter of an hour).  Let $X$ be the number of topping orders in 15 minutes, $X \sim \text{Poisson}(6)$

$$Pr(X = 0) = \frac{6^0 e^{-6}}{0!} \approx 0.0025$$

</details>

## Poisson Distribution in R






#### **R Quick Reference**

`R` supports the following functions for computing distributional quantities and simulating from Poisson distributions:

- `dpois(x, lambda, log = FALSE)` computes PMF of `Poisson(lambda)`;  
- `ppois(q, lambda, lower.tail = TRUE, log.p = FALSE)` computes CDF of `Poisson(lambda)`;  
- `qpois(p, lambda, lower.tail = TRUE, log.p = FALSE)` computes `p`-quantile of `Poisson(lambda)`;  
- `rpois(n, lambda)` simulates `n` random numbers from `Poisson(lambda)`.

Here:

- `x`, `q`: Number of events.  
- `lambda`: Expected number of events.  
- `log`: Whether or not to return log PMF.  
- `lower.tail`: If `TRUE`, return $\Pr(Y \leq x)$; if `FALSE`, return $\Pr(Y > x)$.

b) What is the probability that the contractor is required at least one day in a 5-day working week? You may assume that the number of failures is independent between days.
